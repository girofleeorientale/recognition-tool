Loading the sets

```python
from matplotlib import pyplot as plt
import pandas as pd

columns = ["imgName", "isDefective", "isThereALine"]
df = pd.read_csv("data19april.csv", names=columns)
dfFissures = df.head(10)
print(dfFissures)
dfManges = df.tail(13)
dfManges = dfManges.reset_index()
print(dfManges)
```

A proportion of cracked cookies where a line is detected
(Expected value: with a line)

```python
# count boolean values and create a dataset of it
trueCountFissures = dfFissures.isThereALine.sum()
falseContFissures = len(dfFissures) - trueCountFissures
cols = ['with a line', 'no line']
values = [trueCountFissures, falseContFissures]
data = pd.DataFrame(dict(cols = cols, values = values))

# group values
pie = data.groupby('cols').agg(values = ('values', sum))

# draw a plot
dataPie = pie.plot(kind='pie', y='values', colormap='winter')
dataPie.set_title('Line found in cracked cookies')
```

A proportion of bitten cookies where a line is detected
(Expected value: no line)

```python
# count boolean values and create a dataset of it
trueCountManges = dfManges.isThereALine.sum()
falseContManges = len(dfManges) - trueCountManges
colsM = ['with a line', 'no line']
valuesM = [trueCountManges, falseContManges]
dataM = pd.DataFrame(dict(colsM = colsM, valuesM = valuesM))

# group values
pieM = dataM.groupby('colsM').agg(valuesM = ('valuesM', sum))

# draw a plot
dataPieM = pieM.plot(kind='pie', y='valuesM', colormap='bwr')
dataPieM.set_title('Line found in bitten cookies')
```
