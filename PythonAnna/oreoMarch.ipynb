Loading the first set

```python
from matplotlib import pyplot as plt
import pandas as pd

columns = ["imgName", "mSIndex", "contourArea", "noiselessArea"]
df = pd.read_csv("dataFirst.csv", names=columns)
dfEntiers = df.head(10)
print(dfEntiers)
dfManges = df.tail(10)
dfManges = dfManges.reset_index()
dfInter = df.head(20)
dfFissures = dfInter.tail(10)
dfFissures = dfFissures.reset_index()
print(dfFissures)
```

First set: matchShapes function 

```python
dfEntiers["mSIndex"].plot(label='entiers', color='blue')
dfFissures["mSIndex"].plot(label='fissures', color='red')
dfManges["mSIndex"].plot(label='manges', color='orange')
plt.ylabel('shape index')
plt.xlabel('biscuit number')
plt.legend()
plt.show
```

First set: contour area

```python
dfEntiers["contourArea"].plot(label='entiers', color='blue')
dfFissures["contourArea"].plot(label='fissures', color='red')
dfManges["contourArea"].plot(label='manges', color='orange')
plt.ylabel('area in pixels')
plt.xlabel('biscuit number')
plt.legend()
plt.show
```

This gave us the first version of sorting algorithm (precise from 70 to 90 % for the first set). However, it gave many wrong results at larger dataset (8 out of 14 positive answers).
It is clear from the following graphs.
A technique with euclidian distance is more efficient to detect defective biscuits: 3 wrong decisions in the set of 46 images.
